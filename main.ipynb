```python
import folium
import pandas as pd
```

```python
data=pd.read_csv("C:\datasets\world\worldcities.csv")
data.head()
```

```python
data.columns
```

```python
data.drop(['iso2','iso3'],axis=1,inplace=True)
```

```python
data.city
```

```python
world_map = folium.Map(location=[0, 0], zoom_start=2, tiles="CartoDB positron")
```

```python
from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster().add_to(world_map)

for _, row in data.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=row['city']
    ).add_to(marker_cluster)
```

```python
test_map=folium.Map(location=[0,0],zoom_start=2,tiles="Cartodb Positron")
```

```python
from folium.plugins import FastMarkerCluster
f_marker=FastMarkerCluster(data[['lat', 'lng']].values.tolist()).add_to(test_map)
test_map.save('world_cities_map_fast')
```

```python
import folium

# Create a map centered at a location
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Define a list of coordinates (lat, lon)
route = [
    [28.6139, 77.2090],  # Delhi
    [26.9124, 75.7873],  # Jaipur
    [25.3176, 82.9739]   # Varanasi
]

# Add the line to the map
folium.PolyLine(route, color="blue", weight=2.5, opacity=1).add_to(m)

# Save the map
m.save("route_map.html")
```
